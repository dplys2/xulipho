```python
kq = np.zeros((dd, 6)).T

for j in range(0,6,1):
    measurement = filepath + 'mau'  + str(j+1) + '.jpg'
    print(measurement)
    kq[j] = xulitt(measurement, y[j], goc[j], ghxd, ghxt, dr, ws, ptd = 0)
    print('Hoan thanh ' + str(int((j+1)/6*100)) + ' %' )
    print('\n')
```
