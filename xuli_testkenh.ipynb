```python
### Parameter
ghxd = 2300 # gioi han x duoi
ghxt = 4300 # gioi han x tren
dd = ghxt - ghxd + 1
dr = 5 # do rong chia 2
y = [819, 1031, 1311, 1519, 1799, 2020]
goc = [183.25, 182.75, 181.25, 181.0, 180.0, 179.0]
ws = 3

filepath = 'D:\DataE\\20190721\\MOTT\\'
phan = 'mo'
number = 11
```

```python
kqd = np.zeros((dd, 6)).T
kqb = np.zeros((dd, 6)).T
kq = np.zeros((dd, 6)).T
    
measurement = filepath + 'mau'  + str(number + 1) + '.jpg'
base = filepath + 'mau'  + str(number) + '.jpg'
print(base)
for j in range(0,6,1):
    kqb[j], kqd[j], kq[j] = xuliabs(base, measurement, y[j], goc[j], ghxd, ghxt, dr, ws, ptd = 0)
    print('Hoan thanh ' + str(int((j+1)/6*100)) + ' %' )
```
