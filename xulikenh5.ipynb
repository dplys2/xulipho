```python
#Xu li Methylene blue tu 0.03mM den 0.005mM

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
from skimage import measure
from skimage.transform import rotate
from PIL import Image
import scipy.ndimage

filepath = 'D:\DataE\\20190709\\'
```

```python
#Pre-process Data

fnameb0 = filepath + "0.jpg"
fnameb1 = filepath + "1.jpg"
fnameb2 = filepath + "2.jpg"
fnameb3 = filepath + "3.jpg"
fnameb4 = filepath + "4.jpg"
fnameb5 = filepath + "5.jpg"
fnameb6 = filepath + "6.jpg"
mimg0 = mpimg.imread(fnameb0)
mimg1 = mpimg.imread(fnameb1)
mimg2 = mpimg.imread(fnameb2)
mimg3 = mpimg.imread(fnameb3)
mimg4 = mpimg.imread(fnameb4)
mimg5 = mpimg.imread(fnameb5)
mimg6 = mpimg.imread(fnameb6)
```

```python
### Parameter
ghxd = 1600 # gioi han x duoi
ghxt = 3500 # gioi han x tren
dd = ghxt - ghxd + 1
dr = 20 # do rong chia 2
y1 = 1833
goc12 = 179.5
```

```python
#Cuong do den LED
bxoay1 = rotate(mimg0, goc12)
base_xoay1 = np.dot(bxoay1[...,:3], [299, 587, 114]) #.299 + .587 + .114 
x = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    x[i] = measure.profile_line(base_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))
    
ty = np.mean(x, axis = 0)
```

```python
#Kenh1 va Kenh2
mxoay1 = rotate(mimg1, goc12)
measure_xoay1 = np.dot(mxoay1[...,:3], [299, 587, 114])

n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    n[i] = measure.profile_line(measure_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

tm = np.mean(n, axis = 0)
# ty = np.amax(x, axis = 0)
# tm = np.amax(n, axis = 0)

#np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
c1d = tm
c1 = -np.log10(xm)

del n
del xm

mxoay1 = rotate(mimg2, goc12)
measure_xoay1 = np.dot(mxoay1[...,:3], [299, 587, 114])

n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    n[i] = measure.profile_line(measure_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

tm = np.mean(n, axis = 0)
# ty = np.amax(x, axis = 0)
# tm = np.amax(n, axis = 0)

#np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
c2d = tm
c2 = -np.log10(xm)

del n
del xm

#Kenh1 va Kenh2
mxoay1 = rotate(mimg3, goc12)
measure_xoay1 = np.dot(mxoay1[...,:3], [299, 587, 114])

n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    n[i] = measure.profile_line(measure_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

tm = np.mean(n, axis = 0)
# ty = np.amax(x, axis = 0)
# tm = np.amax(n, axis = 0)

#np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
c3d = tm
c3 = -np.log10(xm)

del n
del xm

mxoay1 = rotate(mimg4, goc12)
measure_xoay1 = np.dot(mxoay1[...,:3], [299, 587, 114])

n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    n[i] = measure.profile_line(measure_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

tm = np.mean(n, axis = 0)
# ty = np.amax(x, axis = 0)
# tm = np.amax(n, axis = 0)

#np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
c4d = tm
c4 = -np.log10(xm)

del n
del xm

mxoay1 = rotate(mimg5, goc12)
measure_xoay1 = np.dot(mxoay1[...,:3], [299, 587, 114])

n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    n[i] = measure.profile_line(measure_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

tm = np.mean(n, axis = 0)
# ty = np.amax(x, axis = 0)
# tm = np.amax(n, axis = 0)

#np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
c5d = tm
c5 = -np.log10(xm)

del n
del xm

mxoay1 = rotate(mimg6, goc12)
measure_xoay1 = np.dot(mxoay1[...,:3], [299, 587, 114])

n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    n[i] = measure.profile_line(measure_xoay1, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

tm = np.mean(n, axis = 0)
# ty = np.amax(x, axis = 0)
# tm = np.amax(n, axis = 0)

#np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
c6d = tm
c6 = -np.log10(xm)
```

```python
plt.plot(c1,'r',label='kenh1')
plt.plot(c2,'g',label='kenh2')
plt.plot(c3,'b',label='kenh3')
plt.plot(c4,'k',label='kenh4')
plt.plot(c5,'c',label='kenh5')
plt.plot(c6,'m',label='kenh6')
plt.legend(loc='upper right')
plt.show()
```
