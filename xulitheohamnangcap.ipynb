```python
### Parameter
ghxd = 1600 # gioi han x duoi
ghxt = 3600 # gioi han x tren
dd = ghxt - ghxd + 1
dr = 5 # do rong chia 2
y = [651, 900, 1203, 1449, 1749, 1986]
goc = [180, 180, 177.75, 177.75, 177, 177]
ws = 49

filepath = 'D:\DataE\\20190714\\MO\\'
phan = 'mo'
```

```python
kqd = np.zeros((dd, 6)).T
kqb = np.zeros((dd, 6)).T
kq = np.zeros((dd, 6)).T

for number in range(1,6,2):
    ad = np.zeros((dd, 6)).T
    ab = np.zeros((dd, 6)).T
    a = np.zeros((dd, 6)).T
    print(base)
    
    measurement = filepath + 'mau'  + str(number + 1) + '.jpg'
    base = filepath + 'mau'  + str(number) + '.jpg'
    for j in range(0,6,1):
        ab[j], ad[j], a[j] = xuli(base, measurement, y[j], goc[j], ghxd, ghxt, dr, ws, ptd = 0)
        print('Hoan thanh ' + str(int((j+1)/6*100)) + ' %' )
    
    print('\n')
    kqd = kqd + ad
    kqb = kqb + ab
    kq = kq + a

kqd = kqd/3.0
kqb = kqb/3.0
kq = kq/3.0
```
