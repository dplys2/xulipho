```python
#Xu li Methylene blue tu 0.03mM den 0.005mM

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
from skimage import measure
from skimage.transform import rotate
from PIL import Image
import scipy.ndimage

filepath = 'D:\DataE\\20190625MO\\'
```

```python
#Pre-process Data

fnameb1 = filepath + "mau1_base1.jpg"
fnameb2 = filepath + "mau2_base1.jpg"
fnameb3 = filepath + "mau3_base1.jpg"
bimg1 = mpimg.imread(fnameb1)
bimg2 = mpimg.imread(fnameb2)
bimg3 = mpimg.imread(fnameb3)

fnameb1 = filepath + "mau1_measure2.jpg"
fnameb2 = filepath + "mau2_measure2.jpg"
fnameb3 = filepath + "mau3_measure2.jpg"
mimg1 = mpimg.imread(fnameb1)
mimg2 = mpimg.imread(fnameb2)
mimg3 = mpimg.imread(fnameb3)
```

```python
#Parameter
ghxd = 3000 # gioi han x duoi
ghxt = 5000 # gioi han x tren
dd = ghxt - ghxd + 1
dr = 10 # do rong chia 2
y1 = 804
y2 = 1050
y3 = 1386
y4 = 1626
y5 = 1986
y6 = 2226
goc12 = 181.75
goc34 = 181.5
goc56 = 179.75
```

```python
#Kenh1 va Kenh2
bxoay1 = rotate(bimg1, goc12)
bxoay2 = rotate(bimg2, goc12)
bxoay3 = rotate(bimg3, goc12)
bxoay = np.sum([bxoay1,bxoay2,bxoay3],axis=0)/3
base_xoay12 = np.dot(bxoay[...,:3], [100, 100, 100])

mxoay1 = rotate(mimg1, goc12)
mxoay2 = rotate(mimg2, goc12)
mxoay3 = rotate(mimg3, goc12)
mxoay = np.sum([mxoay1,mxoay2,mxoay3],axis=0)/3
measure_xoay12 = np.dot(mxoay[...,:3], [100, 100, 100])

x = np.zeros((dr*2,dd))
n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    x[i] = measure.profile_line(base_xoay12, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))
    n[i] = measure.profile_line(measure_xoay12, (y1 - dr + i, ghxd), (y1 - dr + i, ghxt))

ty = np.mean(x, axis = 0)
tm = np.mean(n, axis = 0)

np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
kenh1b = ty
kenh1 = -np.log10(xm)

del x
del n
del xm
x = np.zeros((dr*2,dd))
n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    x[i] = measure.profile_line(base_xoay12, (y2 - dr + i, ghxd), (y2 - dr + i, ghxt))
    n[i] = measure.profile_line(measure_xoay12, (y2 - dr + i, ghxd), (y2 - dr + i, ghxt))

ty = np.mean(x, axis = 0)
tm = np.mean(n, axis = 0)

np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
kenh2b = ty
kenh2 = -np.log10(xm)
```

```python
#Kenh5 va Kenh6
bxoay1 = rotate(bimg1, goc56)
bxoay2 = rotate(bimg2, goc56)
bxoay3 = rotate(bimg3, goc56)
bxoay = np.sum([bxoay1,bxoay2,bxoay3],axis=0)/3
base_xoay12 = np.dot(bxoay[...,:3], [100, 100, 100])

mxoay1 = rotate(mimg1, goc56)
mxoay2 = rotate(mimg2, goc56)
mxoay3 = rotate(mimg3, goc56)
mxoay = np.sum([mxoay1,mxoay2,mxoay3],axis=0)/3
measure_xoay12 = np.dot(mxoay[...,:3], [100, 100, 100])

x = np.zeros((dr*2,dd))
n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    x[i] = measure.profile_line(base_xoay12, (y5 - dr + i, ghxd), (y5 - dr + i, ghxt))
    n[i] = measure.profile_line(measure_xoay12, (y5 - dr + i, ghxd), (y5 - dr + i, ghxt))

ty = np.mean(x, axis = 0)
tm = np.mean(n, axis = 0)

np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
kenh5b = ty
kenh5 = -np.log10(xm)

del x
del n
del xm
x = np.zeros((dr*2,dd))
n = np.zeros((dr*2,dd))
for i in range(0, dr*2):
    x[i] = measure.profile_line(base_xoay12, (y6 - dr + i, ghxd), (y6 - dr + i, ghxt))
    n[i] = measure.profile_line(measure_xoay12, (y6 - dr + i, ghxd), (y6 - dr + i, ghxt))

ty = np.mean(x, axis = 0)
tm = np.mean(n, axis = 0)

np.seterr(divide='ignore', invalid='ignore')
xm = np.divide(tm,ty)
kenh6b = ty
kenh6 = -np.log10(xm)
```
