```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
from skimage import measure
from skimage.transform import rotate
from scipy.signal import savgol_filter
from tvxla import *
```

```python
### Parameter
ghxd = 1600 # gioi han x duoi
ghxt = 3600 # gioi han x tren
dd = ghxt - ghxd + 1
dr = 5 # do rong chia 2
y = [651, 900, 1203, 1449, 1749, 1986]
goc = [180, 180, 177.75, 177.75, 177, 177]
ws = 49

filepath = 'D:\DataE\\20190715\\MOTT\\'
phan = 'mo'
number = 11
```

```python
kqd = np.zeros((dd, 6)).T
kqb = np.zeros((dd, 6)).T
kq = np.zeros((dd, 6)).T
    
measurement = filepath + 'mau'  + str(number + 1) + '.jpg'
base = filepath + 'mau'  + str(number) + '.jpg'
print(base)
for j in range(0,6,1):
    kqb[j], kqd[j], kq[j] = xuliabs(base, measurement, y[j], goc[j], ghxd, ghxt, dr, ws, ptd = 0)
    print('Hoan thanh ' + str(int((j+1)/6*100)) + ' %' )
```

```python
for i in range(0,6,1):
    plt.plot(kq[i], label='kenh' + str(i))
plt.legend(loc='upper right')
plt.show()
```
